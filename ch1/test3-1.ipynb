```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager, rc

# 한글 폰트 설정
plt.rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False

# 데이터 준비
districts = ['강남구', '강동구', '강북구', '강서구', '관악구', '광진구', '구로구', '금천구', 
            '노원구', '도봉구', '동대문구', '동작구', '마포구', '서대문구', '서초구', 
            '성동구', '성북구', '송파구', '양천구', '영등포구', '용산구', '은평구', '종로구', '중구', '중랑구']

crime_counts = [449, 156, 153, 262, 320, 240, 281, 151, 197, 102, 173, 285, 
                294, 154, 393, 126, 150, 220, 120, 295, 194, 166, 211, 170, 187]

# 데이터프레임 생성
df = pd.DataFrame({
    '지역구': districts,
    '범죄 건수': crime_counts
})

# 그래프 크기 설정
plt.figure(figsize=(15, 8))

# 막대 그래프 생성
sns.barplot(data=df, x='지역구', y='범죄 건수', color='skyblue')

# 그래프 제목과 레이블 설정
plt.title('서울시 구별 범죄 발생 현황', fontsize=16, pad=20)
plt.xlabel('지역구', fontsize=12)
plt.ylabel('범죄 건수', fontsize=12)

# x축 레이블 회전
plt.xticks(rotation=45, ha='right')

# 그리드 추가
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 여백 조정
plt.tight_layout()

# 그래프 표시
plt.show()
```
